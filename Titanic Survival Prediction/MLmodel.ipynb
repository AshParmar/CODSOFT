# DataSet Cleaning
#### Time For Dataset cleaning guyz 😎 , removing the null values and columns which are unrelated to our model.



```python
#Calculating the Mean
titanicx["Age"].mean()
```

```python
titanicx.groupby(by="Pclass")["Age"].mean()
```

```python
# 🧑‍🔬 Function to fill missing age values based on passenger class
def age(i):
   Age=i.iloc[0]
   Pclass=i.iloc[1]
   if pd.isnull(Age):
       if Pclass==1:
           return 38
       elif Pclass==2:
           return 29
       elif Pclass==3:
           return 25
   else :
       return Age
    
```

```python
titanicx["Age"]=titanicx.apply(age,axis=1)
```

```python
sns.heatmap(titanicx.isnull(),yticklabels=False)
```

#### Succesfully filled Null value of Ages


```python
# 🗑️ Dropping the 'Cabin' column from the Titanic dataset due to too much null values
titanicx.drop("Cabin",inplace=True,axis=1)
```

```python
sns.heatmap(titanicx.isnull())
```

```python
# 🧑‍💻 Converting the 'Sex' column into dummy/one-hot encoded variables for better model training 
sex=pd.get_dummies(titanicx["Sex"] ,drop_first=True)
sex=sex.astype(int)
sex
```

```python
# 🧳 Converting the 'Embarked' column into dummy/one-hot encoded variables
embark=pd.get_dummies(titanicx["Embarked"],drop_first=True)
embark=embark.astype(int)
embark
```

```python
# 🗑️ Dropping unnecessary columns from the Titanic dataset
titanicx.drop(["Embarked","Sex","Name","PassengerId","Ticket"],axis=1,inplace=True)
```

```python
titanicx
```

```python
#🔗 Concatenating the one-hot encoded 'Sex' and 'Embarked' columns to the Titanic dataset
titanicx=pd.concat([titanicx,sex,embark],axis=1)
titanicx.head()
```

# MODEL BUILDING


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# 🚂 Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(titanicx.drop("Survived", axis=1), 
                                                    titanicx["Survived"], 
                                                    test_size=0.30, 
                                                    random_state=50)
model = LogisticRegression(max_iter=1000)
model.fit(X_train, Y_train)
```

```python
#🔮 Making predictions using the trained Logistic Regression model
prediction=model.predict(X_test)
prediction
```

```python
# 📊 Calculating the accuracy of the Logistic Regression model
accuracy=accuracy_score(Y_test,prediction)
accuracy
```

```python
# Comparing the true values (Y_test) with the predicted values
cm=confusion_matrix(Y_test,prediction)
cm
```
